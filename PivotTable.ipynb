## INTRODUCTION TO PIVOT TABLES IN PANDAS
### Pivot Tables
#### pivot(), pivot_table()

```python
import pandas as pd
```

```python
df_gdp = pd.read_csv('gdp-gdp.csv')
df_gdp
```

```python
#reshape the dataframe with .pivot()
df_gdp.pivot(index="year", columns="country", values="gdppc")
```

#### Pivot_table() method

```python
# read and show the dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales
```

```python
# make a pivot table and add an aggregate function
# Remove datetime columns from the aggregation
numeric_columns = df_sales.select_dtypes(include=['number']).columns

# Create the pivot table with only numeric columns
df_sales.pivot_table(index="Gender", values=numeric_columns, aggfunc="sum")
```

```python
# make a pivot table, and an aggregate function and select some columns
df_sales.pivot_table(index="Gender", values=["Quantity", "Total"], aggfunc="sum")
```

```python
# make a pivot table that says how much male and female spend in each countries
df_sales.pivot_table(index="Gender", columns="Product line", values="Total", aggfunc="sum")
```

# DATA VISUALIZATION WITH PANDAS

```python
# reading the csv file
df_population_raw = pd.read_csv('population_total.csv')
df_population_raw
```

### Making A Pivot Table

```python
# dropping null values
df_population = df_population_raw.dropna()
df_population
```

```python
# making a pivot table
df_pivot = df_population.pivot(index="year", columns="country", values="population")
df_pivot
```

```python
# selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
# showing pivot table
df_pivot
```

## LINEPLOT

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population (1955-2020)', figsize=(8, 4))

# SAVE PLOT
plt.savefig('my_test_png')
# SHOW PLOT
plt.show()
```

```python
df_pivot.to_excel('pivot_table.xlsx')
```

### BARPLOT
##### 1. SINGLE BARPLOT

```python
# selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# showing new dataframe
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar', color='green', xlabel='Year', ylabel='Population', title='Population 2020')
```

#### 2. Barplot grouped by "n" variables

```python
# selecting a few years
df_pivot_Sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020,])]
```

```python
df_pivot_Sample
```

```python
# making grouped barplot
df_pivot_Sample.plot(kind='bar', xlabel='Year', ylabel='Population', title='Population 1980-2020')
```
